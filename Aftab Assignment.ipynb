```python
import numpy as np
```

```python
df_weather = pd.read_csv('weather_forecast_data.csv')
temperature_array = np.array(df_weather['Temperature'])
humidity_array = np.array(df_weather['Humidity'])
print(temperature_array[:5])  
print(humidity_array[:5])  
```

```python
zeros_array = np.zeros((2, 3))  
ones_array = np.ones((2, 3))    
arange_array = np.arange(0, 10, 2)  
print("Zeros Array:")
print(zeros_array)
print("\nOnes Array:")
print(ones_array)
print("\nArange Array:")
print(arange_array)
```

```python
temperature_array = np.array(df_weather['Temperature'])
humidity_array = np.array(df_weather['Humidity'])
sum_arrays = temperature_array + humidity_array
diff_arrays = temperature_array - humidity_array
squared_temperature = temperature_array ** 2
print("Sum of Temperature and Humidity:")
print(sum_arrays[:5])  
print(diff_arrays[:5])  
print("\nSquared Temperature:")
print(squared_temperature[:5])  
```

```python
greater_than_25 = temperature_array > 25
```

```python
temperature_array = np.array(df_weather['Temperature'])
greater_than_25 = temperature_array > 25
print("Temperatures greater than 25 (Boolean Array):")
print(greater_than_25[:10])  
print("\nTemperatures with values greater than 25:")
print(temperature_array[greater_than_25])  
```

```python
temperature_array = np.array(df_weather['Temperature'])
first_ten_temperatures = temperature_array[:10]
print("First 10 Temperatures:")
print(first_ten_temperatures)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df_weather['Temperature'][:50], label='Temperature')
plt.xlabel('Index')
plt.ylabel('Temperature')
plt.title('Temperature Trend')
plt.legend()
plt.show()
```

```python
plt.hist(df_weather['Humidity'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.title('Humidity Distribution')
plt.show()
```

```python
plt.scatter(df_weather['Wind_Speed'], df_weather['Cloud_Cover'], color='red')
plt.xlabel('Wind Speed')
plt.ylabel('Cloud Cover')
plt.title('Wind Speed vs Cloud Cover')
plt.show()
```

```python
rain_counts = df_weather['Rain'].value_counts()
plt.bar(rain_counts.index, rain_counts.values, color=['green', 'blue'])
plt.xlabel('Rain')
plt.ylabel('Count')
plt.title('Rainy vs Non-Rainy Days')
plt.show()
```
